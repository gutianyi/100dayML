## Server

```python
import socket

capital_dic = { "CHINA":"BEIJING",
                "USA":"DC",
                "UK":"LONDON",
                "FRANCE":"PARIS",
                "JAPAN":"TOKYO",
                "INDIA":"DELI",
               }

#hello__list = ["HI", "HELLO", "NICE TO MEET YOU", "NICE TO MEET YOU"]

hello_reply_list = ["Hi~", "Y^_^Y", "Hello!", "So happy talk to u~"]

welcome_list = ["My pleasure", "Let's learn together~", "You are welcome~ ^_^"]

server = socket.socket()
host = socket.gethostname()
port = 8888
server.bind((host, port))

server.listen(1)
while True:
    print("Waiting for client to conect")
    client, addr = server.accept()
    print("Clinet connected, the address is {}".format(addr))
    client.send(bytes("Hello, this is a capital-querying bot.\n May I Know ur name?", encoding = "UTF-8"))
    while True:
        reply = str(client.recv(1024), encoding = "UTF-8")
        re_reply_msg = "Hello " + reply + ", nice to meet you!" + "\n" + "Which country's capital do you want to know about?"
        client.send(bytes(re_reply_msg, encoding = "UTF-8"))
        while True:
            reply = str(client.recv(1024), encoding = "UTF-8")
            print("Client : {}".format(reply))
            reply = reply.upper()
            if reply in capital_dic.keys():
                re_reply_msg = "The capital of " + reply + " is " + capital_dic[reply]
                client.send(bytes(re_reply_msg, encoding = "UTF-8"))
            elif "THANK" in reply and len(reply) <= 10:
                re_reply_msg = choice(welcome_list)
                client.send(bytes(re_reply_msg, encoding = "UTF-8"))
            elif ("HELLO" in reply and len(reply) <= 10) or ("HI" in reply and len(reply) <= 10):
                re_reply_msg = choice(hello_reply_list)
                client.send(bytes(re_reply_msg, encoding = "UTF-8"))
            else:
                re_reply_msg = "Sorry, I don't know. Please let me learn more~" 
                client.send(bytes(re_reply_msg, encoding = "UTF-8"))
        
    client.close()
```

```python
import socket 

client = socket.socket()
host = socket.gethostname()
port = 8080
client.connect((host, port))
while True:
    server_reply = str(client.recv(1024), encoding = "UTF-8")
    print("Server : {}".format(server_reply ))
    msg = input("Client :")
    client.send(bytes(msg, encoding = "UTF-8"))
    while True:
        server_reply = str(client.recv(1024), encoding = "UTF-8")
        print("Server : {}".format(server_reply ))
        msg = input("Client :")
        client.send(bytes(msg, encoding = "UTF-8"))   
```

```python
import pandas as pd
from random import choice


#导入夸夸语料对
kuakua_pairs = []
q_list = []
with open('kuakua_qa_pairs.txt', 'r', encoding='utf-8') as inf:
    for line in inf:
        items = line.strip('\r\n').split('\t')
        kuakua_pairs.append(items)
        if (items[0] in q_list) == False:
            q_list.append((items[0], items[2]))
        
#问句语料切句
import jieba
q_jieba_list = []
for i in range(len(q_list)):
    temp = list(jieba.cut(q_list[i][0]))
    q_jieba_list.append(temp)

#导入词向量模型
from gensim.models import Word2Vec
model = Word2Vec.load("model/word2vec_kuakua.model")

#WMD距离匹配初始化
from gensim.similarities import WmdSimilarity
num_best = 1
instance = WmdSimilarity(q_jieba_list , model, num_best)

query = list(jieba.cut("夸夸我吧"))
sims = instance[query][0][0]
```

```python
candidate_list = []
for i in range(len(kuakua_pairs)):
    if q_list[sims][1] == kuakua_pairs[i][2]:
        candidate_list.append(kuakua_pairs[i][1])
```

```python
server.listen(5)
while True:
    print("Waiting for client to conect")
    client, addr = server.accept()
    print("Clinet connected, the address is {}".format(addr))
    client.send(bytes("你好，我是金圣水傻逼夸夸机器人。请问你是谁？", encoding = "UTF-8"))
    while True:
        reply = str(client.recv(1024), encoding = "UTF-8")
        re_reply_msg = "你好 " + reply + "，你可真是个傻逼。"  + "快让我抹了蜜的小嘴夸夸你吧"
        client.send(bytes(re_reply_msg, encoding = "UTF-8"))
        while True:
            reply = str(client.recv(1024), encoding = "UTF-8")
            print("Client : {}".format(reply))
            #对客户REPLY做的操作
            query = list(jieba.cut(reply))
            sims = instance[query][0][0]
            candidate_list = []
            for i in range(len(kuakua_pairs)):
                if q_list[sims][1] == kuakua_pairs[i][2]:
                    candidate_list.append(kuakua_pairs[i][1])
            re_reply_msg = choice(candidate_list)
            client.send(bytes(re_reply_msg, encoding = "UTF-8"))
```

```python
q_list = []
a_list = []
with open('kuakua_qa_pairs.txt', 'r', encoding='utf-8') as inf:
    for line in inf:
        items = line.strip('\r\n').split('\t')
        q_list.append(items[0])
        a_list.append(items[1])       
```

```python
kuakua_dict = {}
for i in range(len(q_list)):
    if (q_list[i] in kuakua_dict) == False:
        kuakua_dict[q_list[i]] = []
        kuakua_dict[q_list[i]].append(a_list[i])
    if (q_list[i] in kuakua_dict) == True:
        kuakua_dict[q_list[i]].append(a_list[i])
```
